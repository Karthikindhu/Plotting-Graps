```python
# Stacked bar plots

import pandas as pd 
data = {'greenData': [20, 1.5, 7, 10, 5],
        'orangeData': [5, 15, 5, 10, 15],
        'blueData': [2, 15, 18, 5, 10]}

df = pd.DataFrame(data)

total = [i+j+k for i, j, k in zip(df['greenData'], df['orangeData'], df['blueData'])]

green = [(i/j)*100 for i, j in zip(df['greenData'], total)]

orange = [(i/j)*100 for i, j in zip(df['orangeData'], total)]

blue = [(i/j)*100 for i, j in zip(df['blueData'], total)]

name = ['A', 'B', 'C', 'D', 'E']

plt.bar(x_pos, green, color='green')
plt.bar(x_pos, orange, bottom=green, color='orange')
plt.bar(x_pos, blue, bottom=[i+j for i, j in zip(green , orange)], color='blue')

plt.xticks(x_pos, name)

plt.show()
```

```python
# Area plot
x = np.arange(1, 6)

y = [1, 4, 6, 8, 4]

plt.fill_between(x, y)

plt.show()

from random import randint 
x = np.arange(1, 15)

y = [randint(1, 10) for i in range(14)]

plt.fill_between(x, y, color=(0.1, 0.1, 0.1, 0.1))
plt.plot(x, y, color="black")
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")

plt.show()
```

```python
# stacked area plot
y1 = [randint(1, 10) for i in range(14)]
y2 = [randint(1, 10) for i in range(14)]

plt.stackplot(x, y, y1, y2, labels=['A', 'B', 'C'])

plt.legend()

plt.show()
```

```python
# plotly
# Plotly has a hover capability.

!pip install plotly

# plotly.express
import plotly.express as px 

# creating the figure instance
fig = px.line(x=[1, 2, 3], y=[1, 2, 3])

# print the properties of fig
print(fig)

fig.show()

# loading the iris dataset
df_iris = px.data.iris()

df_iris.head(10)

fig = px.line(df_iris, x="species", y="petal_width")

fig.show()

fig = px.line(df_iris, x="sepal_width", y="petal_width")

fig.show()

fig = px.line(df_iris, x="sepal_width", y="petal_width", color="species")

fig.show()
```

```python
# bar plot
long_df = px.data.medals_long()

long_df.head()

fig = px.bar(long_df, x="nation", y="count")
fig.show()

fig = px.bar(long_df, x="nation", y="count", color="medal")
fig.show()

fig = px.bar(long_df, x="nation", y="count", color="medal", text_auto=True)
fig.show()

fig = px.bar(long_df, x="medal", y="count", color="nation", text="nation")
fig.show()

fig = px.bar(long_df, x="medal", y="count", color="nation", pattern_shape="nation", pattern_shape_sequence=[".", "x", "+"])
fig.show()

fig = px.bar(df_iris, x="sepal_width", y="sepal_length")
fig.show()

fig = px.bar(df_iris, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
# Gantt charts

import pandas as pd

df = pd.DataFrame({
    'Task': ["Job A", "Job B", "Job C"],
    'Start': ['2009-01-01', '2009-03-05', '2009-02-20'],
    'Finish': ['2009-02-27', '2009-04-15', '2009-05-30']
})

df
```

```python
# timeline
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task")
fig.show()

df = pd.DataFrame({
    'Task': ["Job A", "Job B", "Job C"],
    'Start': ['2009-01-01', '2009-03-05', '2009-02-20'],
    'Finish': ['2009-02-27', '2009-04-15', '2009-05-30'],
    'TL': ['Amit', 'Omkar', 'Karthik']
})

df

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", color='TL')
fig.show()
```
